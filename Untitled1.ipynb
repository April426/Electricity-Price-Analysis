```python
fig, ax1 = plt.subplots(figsize=(16,8))
ax2 = ax1.twinx()

sns.lineplot(data=energy_df, x="date", y='demand_pos_RRP', ax=ax1, color='tab:blue')  # Plot on ax1
sns.lineplot(data=energy_df, x="date", y='max_temperature', ax=ax2, color='darkorange')  # Plot on ax2

ax1.set_xlabel('Date')
ax1.set_ylabel('Demand with positive RRP', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2.set_ylabel('Max Temperature', color='darkorange')
ax2.tick_params(axis='y', labelcolor='darkorange')


plt.title('Daily Demand with positive RRP VS. Max Temperature')
plt.show()

```

```python
import numpy as np
def find_boundaries(df, variable, distance):

    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)

    lower_boundary = df[variable].quantile(0.25) - (IQR * distance)
    upper_boundary = df[variable].quantile(0.75) + (IQR * distance)

    return upper_boundary, lower_boundary

RRP_upper_limit, RRP_lower_limit = find_boundaries(energy_df, 'RRP', 1.5)
demand_upper_limit, demand_lower_limit = find_boundaries(energy_df, 'demand', 1.5)

outliers_RRP = np.where(energy_df['RRP'] > RRP_upper_limit, True,
              np.where(energy_df['RRP'] < RRP_lower_limit, True, False))

outliers_demand = np.where(energy_df['demand'] > demand_upper_limit, True,
                 np.where(energy_df['demand'] < demand_lower_limit, True, 
                 False))


data_trimmed = energy_df.loc[~(outliers_RRP+outliers_demand)]
```

```python
plt.figure(figsize=(16,8))
sns.regplot(x=data_trimmed['RRP'], y=data_trimmed['demand'])
plt.show()
```
